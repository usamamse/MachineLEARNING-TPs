```python
#Importation des bibliothèques 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Chargement l'ensemble de données
shoes_dataset = pd.read_csv('Shoe prices.csv')
```

```python
#Affichage desinformations sur les données chargé
shoes_dataset.info()
```

```python
#Cette ligne de commande nous permet de retourne les noms de colonnes
shoes_dataset.columns
```

```python
#cette ligne de commande nous permet de retourne le nombre des valeurs nulles dans dataset
shoes_dataset.isnull().sum()
```

```python
#cette ligne de commande retourne des statistiques descriptives pour les colonnes
shoes_dataset.describe()
```

```python
#Indique le nombre de lignes et de colonnes dans dataframe
shoes_dataset.shape
```

```python
# Selectionner aleatoirement un merceau de 4 lignes a partir de dataset
shoes_dataset.sample(4)
```

```python
#on utilise la methode drop pour supprimer la colonne "Model"
shoes_dataset = shoes_dataset.drop('Model', axis=1)
```

```python
#Cette ligne de commande nous permet de retourne les 5 premiers lignes de la dataframe
shoes_dataset.head()
```

```python
#Pour compter le nombre d'occurences de chaque valeur unique dans la collonne Brand
shoes_dataset['Brand'].value_counts()
```

```python
#Cette commande nous permet de visualiser un diagramme a barres a partir des comptages precedente
shoes_dataset['Brand'].value_counts().plot(kind='bar', legend='false')
```

```python
#C'est une amelioration de graphe precedent en ajoutant un titre et des etiquettes
shoes_dataset['Brand'].value_counts().plot(kind='bar', legend='false')
plt.title('Brand Counts')
plt.xlabel('Brand names')
plt.ylabel('Counts')
plt.show()
```

```python
#Convertion de valeurs de la colonne Type en miniscule en utilisant la methode lower()
shoes_dataset['Type'] = shoes_dataset['Type'].str.lower()
```

```python
shoes_dataset['Type'].value_counts()
```

```python
shoes_dataset['Type'].value_counts().plot(kind='bar', legend='false', color = 'green')
plt.title('Type Counts')
plt.xlabel('Type names')
plt.ylabel('Counts')
plt.show()
```

```python
#Creation d'une fonction nommé add_type qui retourne le Type de chausseurs a partir d'une donnée donné
def add_type(inpt):
    if inpt=='casual' or inpt=='fashion' or inpt=='lifestyle' or inpt=='slides' or inpt=='retro' :
        return inpt
    else :
        return 'sport'
shoes_dataset['Type'] = shoes_dataset['Type'].apply(add_type)
```

```python
shoes_dataset['Type'].value_counts()
```

```python
shoes_dataset['Type'].value_counts().plot(kind='bar', legend='false', color = 'green')
plt.title('New Type Counts')
plt.xlabel('Type names')
plt.ylabel('Counts')
plt.show()
```

```python
## Gender
shoes_dataset['Gender'].value_counts()
```

```python
#commande nous permet de visualiser un diagramme circulaire
shoes_dataset['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle = 90)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
## Size
shoes_dataset['Size'] = shoes_dataset['Size'].str.replace('US','')
```

```python
def add_Color(inpt):
    if inpt=='black' or inpt=='white' or inpt=='grey' or inpt=='black/white' or inpt=='pink' :
        return inpt
    else :
        return 'other'
shoes_dataset['Color'] = shoes_dataset['Color'].apply(add_Color)
```

```python
shoes_dataset['Color'].value_counts()
```

```python
shoes_dataset['Color'].value_counts().plot(kind='bar', legend='false', color = 'green')
plt.title('New Color Counts')
plt.xlabel('Color names')
plt.ylabel('Counts')
plt.show()
```

```python
## Material
shoes_dataset['Material'] = shoes_dataset['Material'].str.lower()
```

```python
shoes_dataset['Material'].value_counts()
```

```python
shoes_dataset['Material'].value_counts().plot(kind='bar', legend='false',  color= 'green')
plt.title('Material Counts')
plt.xlabel('Material names')
plt.ylabel('Counts')
plt.show()
```

```python
def add_Material(inpt):
    if inpt=='mech' or inpt=='leather' or inpt=='canvas' or inpt=='primeknit' or inpt=='synthetic' or inpt=='suede' or inpt=='knit' or inpt=='nylon' :
        return inpt
    else :
        return 'other'
shoes_dataset['Material'] = shoes_dataset['Material'].apply(add_Material)
```

```python
shoes_dataset['Material'].value_counts()
```

```python
shoes_dataset['Material'].value_counts().plot(kind='bar', legend='false',  color= 'green')
plt.title('New Material Counts')
plt.xlabel('Material names')
plt.ylabel('Counts')
plt.show()
```
